```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class DAAMLoss(nn.Module):
    def __init__(self, num_classes, embedding_size=512, s=30, m_min=0.2, m_max=0.5):
        super().__init__()
        self.W = nn.Parameter(torch.randn(num_classes, embedding_size))
        nn.init.xavier_uniform_(self.W)
        self.s = s
        self.m_min = m_min
        self.m_max = m_max

    def forward(self, embeddings, labels, quality):
        W = F.normalize(self.W, dim=1)
        cos_theta = torch.matmul(embeddings, W.t()).clamp(-1+1e-7, 1-1e-7)

        m = self.m_min + quality * (self.m_max - self.m_min)
        theta = torch.acos(cos_theta)
        target_logits = torch.cos(theta + m)

        one_hot = F.one_hot(labels, num_classes=W.size(0)).float()
        logits = cos_theta * (1 - one_hot) + target_logits * one_hot
        logits *= self.s

        return F.cross_entropy(logits, labels)
```

```python
import torch.nn.functional as F
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

def evaluate(pairs, targets, threshold=0.5):
    correct, FAR, FRR = 0, 0, 0
    pos = sum(targets)
    neg = len(targets)-pos

    for (e1,e2), t in zip(pairs, targets):
        sim = F.cosine_similarity(e1.unsqueeze(0), e2.unsqueeze(0)).item()
        pred = 1 if sim>threshold else 0

        if pred==t: correct+=1
        else:
            if t==1: FRR+=1
            else: FAR+=1

    return correct/len(targets), FAR/neg, FRR/pos


def plot_roc(pairs, targets):
    sims = [F.cosine_similarity(e1.unsqueeze(0), e2.unsqueeze(0)).item() for e1,e2 in pairs]
    fpr, tpr, _ = roc_curve(targets, sims)
    plt.plot(fpr, tpr)
    plt.title("ROC Curve")
    plt.xlabel("FAR")
    plt.ylabel("TPR")
    plt.show()
```

```python
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
import torchvision.models as models
import torch.nn.functional as F
from torch.utils.data import DataLoader
import random
```

```python
def get_loader(path, batch_size=32):
    transform = transforms.Compose([
        transforms.Resize((112,112)),
        transforms.ToTensor(),
        transforms.Normalize([0.5]*3, [0.5]*3)
    ])

    dataset = torchvision.datasets.ImageFolder(path, transform=transform)

    loader = DataLoader(
        dataset,
        batch_size=batch_size,
        shuffle=True,
        num_workers=2
    )

    return loader, len(dataset.classes)
```

```python
class FaceModel(nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        base = models.resnet50(pretrained=True)
        self.features = nn.Sequential(*list(base.children())[:-1])
        self.fc = nn.Linear(2048, 512)

        self.classifier = nn.Linear(512, num_classes)

    def forward(self, x):
        x = self.features(x)
        x = x.view(x.size(0), -1)
        emb = F.normalize(self.fc(x), dim=1)

        logits = self.classifier(emb)
        return emb, logits
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'

train_loader, num_classes = get_loader("/content/data/test")  # using your dataset

model = FaceModel(num_classes).to(device)

optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9)
criterion = nn.CrossEntropyLoss()
```

```python
epochs = 5

for epoch in range(epochs):
    model.train()
    total_loss = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()

        emb, logits = model(images)
        loss = criterion(logits, labels)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")
```

```python
model.eval()

embeddings = []
labels_list = []

with torch.no_grad():
    for images, labels in train_loader:
        images = images.to(device)

        emb, _ = model(images)
        embeddings.append(emb.cpu())
        labels_list.append(labels)

embeddings = torch.cat(embeddings)
labels_list = torch.cat(labels_list)

print("Embeddings shape:", embeddings.shape)
```

```python
def create_pairs(embeddings, labels, n=500):
    pairs, targets = [], []
    size = len(embeddings)

    for _ in range(n):
        i, j = random.randint(0,size-1), random.randint(0,size-1)
        pairs.append((embeddings[i], embeddings[j]))
        targets.append(1 if labels[i]==labels[j] else 0)

    return pairs, targets

pairs, targets = create_pairs(embeddings, labels_list)
```

```python
def evaluate(pairs, targets, threshold=0.5):
    correct = 0

    for (e1,e2), t in zip(pairs, targets):
        sim = F.cosine_similarity(e1.unsqueeze(0), e2.unsqueeze(0)).item()
        pred = 1 if sim > threshold else 0

        if pred == t:
            correct += 1

    return correct / len(targets)

accuracy = evaluate(pairs, targets)

print("Final Accuracy:", accuracy)
```
